### Creating data frame manually

```python
# Data frame from a list
height = [180,200,230,280,298,268]
df_list = pd.DataFrame(height)
print(df_list)
```

```python
# Data frame from a list
height = [180,200,230,280,298,268]
df_list = pd.DataFrame(height,columns = ['Height'], index = [10,20,30,40,50,60])
print(df_list)
```

### Filtering data using Masks

```python
# Filtering 
# single = is assignment operator x = 5 
# double == testing the condition output True or False
df[df.salary == 'low']
```

```python
df.salary == 'low'
```

```python
# Satisfaction level more than 0.5
df[df.satisfaction_level > 0.5]
```

```python
# Satisfaction level more than 0.5
df[df.satisfaction_level > 0.5].head()
```

```python
# Multiple condition filtering
df[(df.satisfaction_level > 0.5) & (df.salary == 'low')]
```

```python
# Filtering rows and indexing columns using explicit label
df.loc[df.satisfaction_level > 0.5, ['salary', 'sales', 'satisfaction_level']]
```
