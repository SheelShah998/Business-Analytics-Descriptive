#### Merging movies and ratings


```python
# merging datasets
ratings = pd.merge(movies, ratings).drop(['genres', 'timestamp'], axis = 1)
```

```python
# first few rows
ratings.head()
```

```python
# Info dataframe
ratings.info()
```

#### New Dataframe of UserID by movies 

```python
# creating UserID by movies dataframe
ratings = ratings.pivot_table(index = ['userId'], columns = ['title'], values = 'rating')
```

```python
# Ratings first few rows
ratings.head()
```

```python
# dataframe dimension
ratings.shape
```

```python
# Dropping columns with few ratings
ratings = ratings.dropna(thresh = 10, axis = 1)
```

```python
# Dataframe dimension 
ratings.shape
```

```python
# Head of the first few rows 
ratings.head()
```

```python
# Dropping columns with few ratings
ratings = ratings.dropna(thresh = 20, axis = 1)
```

```python
# Dataframe dimension 
ratings.shape
```

```python
# Head of the first few rows 
ratings.head()
```

#### Item based Collaborative Filtering

```python
# Correlation Matrix
corrMatrix = ratings.corr()
corrMatrix.head()
```

```python
# Find similar movies
corrMatrix['2001: A Space Odyssey (1968)'].sort_values(ascending = False)
```

```python
# Find similar movies
corrMatrix['Young Frankenstein (1974)'].sort_values(ascending = False)
```

```python
# Find similar movies
corrMatrix['13 Going on 30 (2004)'].sort_values(ascending = False)
```
